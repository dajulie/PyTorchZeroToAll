```python
x_data = [1.0, 2.0, 3.0]
y_data = [2.0, 4.0, 6.0]

w=1.0

def forward(x):
  return x * w

def loss(x, y):
  y_pred = forward(x)
  return (y_pred - y) * (y_pred - y)

def gradient(x, y):
  return 2 * x * (x * w - y)

print("predict before training: ", 4, forward(4))

for epoch in range (100):
  for x_val, y_val in zip(x_data, y_data):
    grad = gradient(x_val, y_val)
    w=w-0.01*grad
    print("\tgrad: ", x_val, y_val, grad)
    l=loss(x_val, y_val)

  print("progress: ", epoch, "w= ", w, "loss= ", l)

print("predict after training", "4 hours", forward(4))
```

```python
x_data = [1.0, 2.0, 3.0]
y_data = [2.0, 4.0, 6.0]

w1=1.0
w2=1.0
b=1.0

def forward(x):
  return x**2 * w2 + x * w1 + b

def loss(x, y):
  y_pred = forward(x)
  return (y_pred - y) * (y_pred - y)

def gradient1(x,y):
  return 2*x*(x*w1-y+w2*x**2+b)

def gradient2(x,y):
  return 2 * x**2 * (w2*x**2-y+w1*x+b)

def gradient3(x,y):
  return 2 * (w2*x**2-y+w1*x+b)

print("predict before training: ", 4, forward(4))

for epoch in range (100):
  for x_val, y_val in zip(x_data, y_data):
    grad1 = gradient1(x_val, y_val)
    w1=w1-0.01*grad1
  for x_val, y_val in zip(x_data, y_data):
    grad2 = gradient2(x_val, y_val)
    w2=w2-0.01*grad2
  for x_val, y_val in zip(x_data, y_data):
    grad3 = gradient3(x_val, y_val)
    b=b-0.01*grad3
    print("\tgrad: ", x_val, y_val, grad1, grad2)
    l=loss(x_val, y_val)

  print("progress:",epoch,"w1=",w1,"w2=",w2,"b=", b, "loss=",l)

print("predict after training", "4 hours", forward(4))
```
